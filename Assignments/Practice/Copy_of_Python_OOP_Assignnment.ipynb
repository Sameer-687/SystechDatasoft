### Task 1: Vehicle Inheritance
Create a **base class** called Vehicle with attributes **`manufacturer`** and **`model`**. Implement a method **`get_info()`** that returns a string containing the manufacturer and model of the vehicle. Then, create two derived classes(Inherited Classes), **Car and Motorcycle**, that inherit from Vehicle. Each derived class should have an additional attribute and override the **`get_info()`** method to include the new attribute's value. Finally, create instances of both classes and print their information.

Task 1 solve

```python
class Vehicle:
    def __init__(self,manufacturer,model):
        self.manufacturer=manufacturer
        self.model=model

    def get_info(self):
        return "Manufacturer of the vehicle: {} \nThe model of the vehicle: {}".format(self.manufacturer,self.model)
        #return "Manufacturer of the vehicle: {0.manufacturer} \n The model of the vehicle: {0.model}".format(self)

v1=Vehicle("Mercedes", "CLA")
v2=Vehicle("BMW", "X3")


class Car(Vehicle):
    def __init__(self,manufacturer,model,color):
        super().__init__(manufacturer,model)
        self.color=color
    def get_info(self):
        return "This is a Car.\nManufactured by: {} \nModel of the car: {}\nColor of the car: {} ".format(self.manufacturer,self.model,self.color)

c1=Car("Ford","Mustang","Black")
c2=Car("BMW","X3","White")

class Motorcycle(Vehicle):

    def __init__(self,manufacturer,model,color):
        super().__init__(manufacturer,model)
        self.color=color
    def get_info(self):
        return "This is a Motorcycle.\nManufactured by: {} \nModel of the Bike: {}\nColor of the Bike: {}".format(self.manufacturer,self.model,self.color)

m1=Motorcycle("BMW","1000 XR","Yellow")
print(v1.get_info())
print(c1.get_info())
print(m1.get_info())






```

### Task 2: Online Shopping System

- Create a class named **`Product`** with attributes like **`name`**, **`price`**, and **`quantity_in_stock`**.
<br>

- Create a class named **`Cart`** that allows adding and removing products. It should also calculate the total cost of items in the cart.
<br>

- Implement a class named **`Customer`** with attributes like **`name`**, **`email`**, and a **`cart instance`**.
    Allow the customer to add products to their cart, calculate the final cost, and proceed to checkout.

```python
class Product:
    def __init__(self,name,price,quantity_in_stock):
        self.name=name
        self.price=price
        self.quantity_in_stock=quantity_in_stock


class Cart:
    def __init__(self):
        #super().__init__(name,price,quantity_in_stock):
        self.items=[]

    def add_product(self,product,quantity):

        if quantity>0 and quantity<=product.quantity_in_stock:
            self.items.append([product,quantity])
            product.quantity_in_stock-=quantity
            print(f"{quantity} {product.name}'s  adedd to cart. {product.quantity_in_stock} {product.name} in in stock right now.")

        else:
            print(f"{product.name} is not added in the cart.")

    def remove_product(self,product):
        found=False
        for item in self.items:
            if item[0]==product:
                self.items.remove(item)
                product.quantity_in_stock+=item[1]
                print(f"{product.name} is removed. {product.quantity_in_stock} {product.name} is in the stock. ")
                found=True
                break
        
        if not found:
            print(f"{product.name} is not in the cart.")
            
    def calculate_total_cost(self):
        total=0
        for item in self.items:
            multiplication=item[1]*item[0].price
            total+=multiplication
        return total

class Customer:
    def __init__(self,name,email):
        self.name=name
        self.email=email
        self.cart=Cart()

    def add_to_cart(self,product,quantity):
        self.cart.add_product(product,quantity)

    def remove_from_the_cart(self,product):
        self.cart.remove_product(product)

    def checkout(self):
        cost=self.cart.calculate_total_cost()
        print(f"Thank you for shopping with us. {self.name}!!")
        print(f"Your final cost is ${cost}")
    

#creating product
Laptop=Product('Laptop',800,150)
Mobile=Product("Mobile",350,500)
Tab=Product("Tab",550,300)
Watch=Product("Watch",80,100)

#creating customer
customer1=Customer("David", "David@gmail.com")
customer2=Customer("Miller", "Miller@gmail.com")

#adding product to the cart
customer2.add_to_cart(Laptop,200)
customer2.add_to_cart(Tab,2)
customer2.add_to_cart(Watch,2)
customer2.add_to_cart(Mobile,1)

#removing product
customer2.remove_from_the_cart(Tab)

#checkout
customer2.checkout()



        
        
        


```

### Task 4: Online Banking System

- Create a class **`BankAccount`** with attributes like **`account_number, balance, and account_holder`**.
<br>

- Implement **methods** for depositing, withdrawing, and transferring money between accounts.
<br>

- Create a **subclass** **`SavingsAccount`** that offers an interest rate and has a limit on withdrawals.
<br>

Show how a customer can open a bank account, perform transactions, and earn interest on savings.

```python
class BankAccount:
    def __init__(self,account_number,account_holder,balance):
        self.account_number=account_number
        self.account_holder=account_holder
        self.balance=balance

    def deposit_money(self,amount):
        self.balance+=amount
        print(f"{amount} BDT has been depositted to \nA/C Name: {self.account_holder}\nA/C No: {self.account_number}\nA/C Balance: {self.balance}")

    def withdraw_money(self,amount):
        if amount>self.balance:
            print(f"Insufficient balance! Try to withdraw less than {self.balance} BDT. ")
        else:
            self.balance-=amount
            print(f"{amount} BDT has been withdrawn from \nA/C Name: {self.account_holder}\nA/C No: {self.account_number}\nA/C Balance: {self.balance}")
    def transfer_money(self,target_account,amount):
        if amount>self.balance:
            print(f"Insufficient balance! Try to transfer less than {self.balance} BDT. ")
        else:
            self.balance-=amount
            target_account.balance = int(target_account.balance) + amount
            print(f"{amount} BDT has been transferred to A/C No: {target_account.account_number} from {self.account_number}")
            print(f"Transaction: \n From \nA/C Name: {self.account_holder}, A/C No: {self.account_number}, Balnace: {self.balance}\nTo: \nA/C Name: {target_account.account_holder}, A/C No: {target_account.account_number}, Balance: {target_account.balance}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance,interest_rate, withdrawal_count=0,withdrawal_limit=5):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate
        self.withdrawal_limit = withdrawal_limit
        self.withdrawal_count=0

    def earn_interest(self):
        interest_earned = self.balance * self.interest_rate
        self.balance += interest_earned
        return f"Earned interest: {interest_earned}. New balance: {self.balance}"
    def  withdraw_money(self, amount):
        if self.withdrawal_count<=self.withdrawal_limit:
            result=super().withdraw_money(amount)
            self.withdrawal_count+=1
            return result
        else:
            return "Withdrawal limit reached for this month."
        

    
account_1=BankAccount(123456,"Lionel",10000)
account_2=BankAccount(234567,"Messi",50000)
account_3=SavingsAccount(345678,"Cristiano",30000,0.01)
account_4=SavingsAccount(456789,"Ronaldo",15000,0.02)

account_2.withdraw_money(10000)
account_2.deposit_money(5000)
account_2.transfer_money(account_4,10000)
account_4.earn_interest()



        
```
